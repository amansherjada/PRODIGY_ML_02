```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']= (10,5)
plt.rcParams['figure.dpi']= 250
%matplotlib inline
import seaborn as sns
sns.set_style('darkgrid')
import warnings
warnings.filterwarnings('ignore')
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# Load dataset
data = pd.read_csv('Mall_Customers.csv', index_col=0)
```

```python
data.head()
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
sns.scatterplot(x= data['Annual Income (k$)'], y = data['Spending Score (1-100)'])
plt.show()
```

```python
# Select relevant features for clustering
df = data[['Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
# Standarization
sc = StandardScaler()
x = sc.fit_transform(df)
x
# x stores the standardized data  
```

```python
# Model Building
df.head()
```

```python
km = KMeans()
ypred = km.fit_predict(x)
ypred
```

```python
len(ypred)
```

```python
# cluster center
km.cluster_centers_
```

```python
x[:,0] # Scaled value of our dataset x coordinate
```

```python
x[:,1] # Scaled value of our dataset y coordinate
```

```python
# Visualize the Clusters
plt.scatter(x[:,0], x[:,1], c = ypred, cmap='rainbow') # Clusters
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1], c= 'black', s=100) # Centriods
plt.tight_layout()
plt.show()
```

```python
km.inertia_ # wcss value
```

```python
# Elbow Method
wcss = []
for i in range(1, 11):
    km= KMeans(n_clusters=i)
    km.fit_predict(x)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11), wcss)
plt.axvline(x = 5, color = 'red', linestyle = '--')
plt.xlabel('K-Values')
plt.ylabel('WCSS Value')
plt.title('Elbow Graph')
plt.tight_layout()
plt.show()
```

```python
# Final Model
km = KMeans(n_clusters=5)
ypred = km.fit_predict(x)
ypred
```

```python
# Visualize the Clusters
km.cluster_centers_
```

```python
plt.scatter(x[:,0], x[:,1], c= ypred, cmap="rainbow") # Clusters
# c in below code is color and s is size
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], c ='purple', s= 100) # Centriod
plt.tight_layout()
plt.show()
```

```python
# Evaluation of K-Means Clustering
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(x, ypred)
```
